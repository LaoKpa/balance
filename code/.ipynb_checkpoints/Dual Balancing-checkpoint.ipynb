# Scenario 1 : NID Demand

```r
history <- rnorm(100, 100, 15) + cumsum(rnorm(100, 1, 1))
model <- Arima(history, order=c(0, 0, 0), include.drift=T)
q <- 100
h <- rep(1, 100)
p <- rep(5, 100)
trace <- c()
for (j in 10) {
for (i in 1:100) {
    demand <- simulate(model, nsim=100)
    q <- q - (1/(i+1))*pr_root(q, 0, demand, h, p)
    trace <- c(trace, q)
}
}
plot(trace, pch=".")
lines(trace)
```
